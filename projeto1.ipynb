Hipótese: Com a criação da união africana, os países membros provavelmente obtiveram uma melhora no índice de investimentos, e, como parte desse dinheiro provavelmente foi convertido em melhorias na qualidade do saneamento básico, provavelmente a expectativa de vida aumentou.

Mecanismo: A União Africana tem como objetivo promover a democracia, direitos humanos e o desenvolvimento econômico. Com isso, foi criado um programa que incentiva o aumento dos investimentos de países extrangeiros  na africa. Por isso, junto com as políticas de proteção dos direitos humanos, se realmente houve um aumento de investimentos, a economia de ter melhorado, e, consequentemente, o governo promoveu uma melhora no saneamento básico. 
Como o saneamento básico ajuda a diminuir a proliferação de doenças, a expectativa de vida também deve ter aumentado.
