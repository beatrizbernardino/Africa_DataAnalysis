* **Pergunta**: depois de entrar para a OCDE, o índice de desemprego do país membro diminuiu significamente quando comparado com um país não membro?

* **Hipótese** : Ao entrar na OCDE, há um maior incentivo na área industrial e comercial, gerando novos empregos. 

* **Mecanismo** : Como a OCDE apoia a economia de mercado, que defende a privatização da maior parte da produção economica, os novos países membros seriam alvos de  novos investimentos, gerando novos negócios e expansão dos antigos, demandando portanto, mais trabalhadores. Com isso, se a taxa de desemprego diminuir, consequentemente a o pib per capita deve aumentar, provando que a privatização pode ser uma boa solução para o crescimento econômico.

```python
pibindex=pib.set_index('country')
pibindex.head(3)
```

```python
desindex=desemprego.set_index('country')
desindex.head(3)
desindex.dropna()
```

```python
#tirando os países que não contém dados de desmprego suficientes
des=desindex[(desindex[2009]>0.0)& (desindex[2008]>0.0) & (desindex[2007]>0.0) & (desindex[2006]>0.0) &(desindex[2007]>0.0)]

#separando 3 anos antes dos países entrarem na ocde
a=des.iloc[:, 15:20]
des_paises_antes_ocde=a.transpose()
des_paises_antes_ocde.describe()
media_desemprego=des_paises_antes_ocde.mean()
desmedia = pd.DataFrame(data=media_desemprego, columns = ['media_desemprego'])
desmedia.head(3)
```

```python
#tirando os países que não contém dados de desmprego suficientes
des_ocde=desindex[(desindex[2010]>0.0)& (desindex[2011]>0.0) & (desindex[2012]>0.0) & (desindex[2013]>0.0) &(desindex[2014]>0.0)]

#separando 3 anos antes dos países entrarem na ocde
z=des_ocde.iloc[:, 20:25]

des_paises_depois_ocde=z.transpose()
media_desemprego=des_paises_depois_ocde.mean()
desmedia_ocde = pd.DataFrame(data=media_desemprego, columns = ['media_desemprego'])
desmedia_ocde.head(3)
```

```python
do = desmedia.join(desmedia_ocde, how='inner', lsuffix='_antes', rsuffix='_depois')
comparacao_des=do.iloc[(do.index=='Israel') |(do.index=='Estonia')]
comparacao_des
```
